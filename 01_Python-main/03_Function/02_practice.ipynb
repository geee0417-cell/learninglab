```python
##나만의 계산기 만들기

#1.while 문을 통해서 문한반복, break로 탈출 !
#2.사용자의 입력으로 사용할 함수를 고를 것 (input)
#3.함수만들기?

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "오류: 0으로 나눌 수 없습니다."
    return x / y


def calculator():
    print("---------------------------------")
    print("지안 계산기에 오신 것을 환영합니다")
    print("사용 가능한 연산:")
    print("1. 덧셈 (+)")
    print("2. 뺄셈 (-)")
    print("3. 곱셈 (*)")
    print("4. 나눗셈 (/)")
    print("5. 종료")
    print("---------------------------------")

    while True: # 무한반복
        operation = input("연산자 번호를 입력하세요 (1/2/3/4/5): ")

        # 1. 종료 기능 (가장 먼저 처리)
        if operation == '5':
            print("계산기를 종료합니다. 안녕히 가세요! 👋")
            break

        # 2. 유효한 연산자 (1, 2, 3, 4)인 경우 처리
        elif operation in ('1', '2', '3', '4'):
            try:
                # 숫자 입력 및 float 변환
                num1_input = input("첫 번째 숫자 입력: ")
                num2_input = input("두 번째 숫자 입력: ")
                num1 = float(num1_input) 
                num2 = float(num2_input)

                # 연산 기호 매핑 (출력용)
                op_map = {'1': '+', '2': '-', '3': '*', '4': '/'}
                op_symbol = op_map.get(operation, '?')

                # 계산 로직
                if operation == '1':
                    result = add(num1, num2)
                elif operation == '2':
                    result = subtract(num1, num2)
                elif operation == '3':
                    result = multiply(num1, num2)
                elif operation == '4':
                    result = divide(num1, num2)

                # 결과 출력 (이후 while 루프가 다음 계산을 위해 자동으로 재시작됨)
                print(f"결과: {num1} {op_symbol} {num2} = {result}\n")
            
            except ValueError:
                print("⚠️ 오류: 숫자를 입력해 주세요.\n")
                continue # 다음 루프로 이동하여 연산자 재입력

        # 3. 잘못된 입력 처리 (1~5 외의 모든 입력)
        else:
            print("❌ 잘못된 입력입니다. 1에서 5 사이의 숫자를 입력해 주세요.\n")

# 계산기 실행
calculator()
```
