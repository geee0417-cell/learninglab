1.데이터 읽어오기


```python
import pandas as pd

df_m = pd.read_excel("data/2509연령별인구현황.xlsx", skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m
```

2. 데이터를 문자열에서 정수로 변환해주기

```python
print(df_m.dtypes) #현재문자열
#문자열로변환 -> replace()함수를 통해 콤마(,) 제거 -> 타입변경(int)

df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
```
