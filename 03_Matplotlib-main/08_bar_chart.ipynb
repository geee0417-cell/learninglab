# 07. 막대 그래프(Bar Chart)
---
### 막대 그래프의 활용
- 막대 그래프는 범주형 데이터를 비교할 때 효과적이다.
- 각 항목의 크기를 직관적으로 비교할 수 있어 데이터 분석에서 자주 사용된다.


```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

## 기본 막대 그래프
### bar() 함수
- **bar()** 함수를 사용하여 세로 막대 그래프를 그릴 수 있다.
- 첫 번째 인자는 카테고리(레이블), 두 번째 인자는 값이다.


```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명
plt.bar(cities, population)
plt.title('도시별인구')
plt.ylabel('인구(만명)')

plt.show()
```

## 막대 색상 설정
### color 파라미터
- **color** 파라미터로 막대의 색상을 변경할 수 있다.
- 단일 색상 또는 각 막대마다 다른 색상을 지정할 수 있다.


```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명
plt.bar(cities, population, color='red')
plt.title('도시별인구')
plt.ylabel('인구(만명)')

plt.show()
```

```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명

colors = ['red','green','blue','pink']
plt.bar(cities, population, color=colors)
plt.title('도시별인구')
plt.ylabel('인구(만명)')

plt.show()
```

- 범위조정을 적용시켜보자


```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명

colors = ['red','green','blue','pink']
plt.bar(cities, population, color=colors)
plt.title('도시별인구')
plt.ylabel('인구(만명)')
plt.ylim(200,1000)

plt.show()
```

## 가로 막대 그래프
### barh() 함수
- **barh()** 함수를 사용하여 가로 막대 그래프를 그릴 수 있다.
- 항목이 많거나 라벨이 길 때 유용하다.


```python

plt.barh(cities, population)

plt.title('도시별인구')
plt.xlabel('인구(만명)')
```

## 막대에 값 표시하기
### text() 함수 활용
- 막대 위에 값을 표시하면 데이터를 더 명확하게 전달할 수 있다.
- **get_height()** 메서드로 각 막대의 높이를 가져올 수 있다.


```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명

colors = ['red','green','blue','pink']
plt.title('도시별인구')
plt.ylabel('인구(만명)')
plt.ylim(200,1200)

#각 막대 위에 값을 표시 
bar = plt.bar(cities, population, color=colors)

for idx,rect in enumerate(bar):
    #축의 높이를 가져올 수 있음
    height = rect.get_height()
    plt.text(idx, height+20, f'{population[idx]}만명', ha= 'center', color=colors[idx])

plt.show()
```

## 막대 패턴 설정
### set_hatch() 메서드
- 막대에 패턴을 추가하면 흑백 인쇄 시에도 막대를 구분할 수 있다.
- 다양한 패턴: '/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*'


```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명

colors = ['red','green','blue','pink']
plt.title('도시별인구')
plt.ylabel('인구(만명)')
plt.ylim(200,1200)

#각 막대 위에 값을 표시 
bar = plt.bar(cities, population, color=colors)

for idx,rect in enumerate(bar):
    #축의 높이를 가져올 수 있음
    height = rect.get_height()
    plt.text(idx, height+20, f'{population[idx]}만명', ha= 'center', color=colors[idx])

bar[0].set_hatch('*') #바꾸미기
bar[1].set_hatch('-')
bar[2].set_hatch('/')
bar[3].set_hatch('O')

plt.show()
```

```python
cities=['서울','대구','부산','창원']
population= [970,342,242,295] #단위 : 만명

colors = ['red','green','blue','pink']
hatch = ['*','-','O','X']
plt.title('도시별인구')
plt.ylabel('인구(만명)')
plt.ylim(200,1200)

#각 막대 위에 값을 표시 
bar = plt.bar(cities, population, color=colors)

for idx,rect in enumerate(bar):
    #축의 높이를 가져올 수 있음
    height = rect.get_height()
    plt.text(idx, height+20, f'{population[idx]}만명', ha= 'center', color=colors[idx])

    bar[idx].set_hatch(hatch[idx])

plt.show()
```
