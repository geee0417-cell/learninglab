# 15. 산점도 (Scatter Plot)
---
### 산점도의 특징
- 산점도는 두 변수 간의 관계를 시각화하는 데 효과적이다.
- 데이터의 분포와 상관관계를 한눈에 파악할 수 있다.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

## 데이터 준비
### 고객 데이터 불러오기
- customers3.csv 파일에 학년 정보를 추가한다.


```python
# 고객 데이터 불러오기
df = pd.read_csv('../02_Pandas-main/data/customers3.csv', encoding='utf-8')
df.head()
```

## 기본 산점도
### scatter() 함수
- **scatter()** 함수를 사용하여 산점도를 그린다.
- X축과 Y축에 각각의 변수를 지정한다.


```python
plt.scatter(df['평균구매액(만원)'],df['고객만족도(점)']) #두 인자간의 관계확인
plt.xlabel('평균구매액(만원)')
plt.ylabel('고객만족도(점)')
plt.show()
```

## 점 크기 조정하기
### s 파라미터
- **s** 파라미터로 점의 크기를 조정할 수 있다.
- 다른 변수의 값에 따라 점의 크기를 다르게 할 수 있다.


```python
sizes= df['나이'] #나이에 따라 크기가 달라지게함
plt.scatter(df['평균구매액(만원)'],df['고객만족도(점)'],s=sizes*20) #두 인자간의 관계확인
plt.xlabel('평균구매액(만원)')
plt.ylabel('고객만족도(점)')
plt.show()
```

## 점 색상 조정하기
### c와 cmap 파라미터
- **c**: 점의 색상을 지정한다.
- **cmap**: 컬러맵을 지정한다.
    - viridis : 파란색-녹색-노란색 으로이루어진 색상 맵맵=
    - 변수 값에 따라 색상이 자동으로 매핑된다.


```python
sizes= df['나이'] #나이에 따라 크기가 달라지게함

plt.scatter(df['평균구매액(만원)'],df['고객만족도(점)'],s=sizes*20,  cmap= 'viridis',c=df['재구매의사(점)']) 
plt.xlabel('평균구매액(만원)')
plt.ylabel('고객만족도(점)')
plt.show()
```

## 컬러바 추가하기
### colorbar() 함수
- **colorbar()**로 색상과 값의 관계를 보여주는 컬러바를 추가한다.
- **label**: 컬러바 레이블
- **shrink**: 컬러바 크기 조정
- **orientation**: 컬러바 방향 ('vertical', 'horizontal')


```python
sizes= df['나이'] #나이에 따라 크기가 달라지게함

plt.scatter(df['평균구매액(만원)'],df['고객만족도(점)'],s=sizes*20, c=df['고객만족도(점)'])

plt.colorbar(label='고객만족도(점)',shrink=0.7)

plt.xlabel('평균구매액(만원)')
plt.ylabel('고객만족도(점)')
plt.show()
```
