# 11. 누적 막대 그래프

### 누적 막대 그래프의 활용
- 누적 막대 그래프는 여러 데이터 계열을 하나의 막대에 쌓아서 표현한다.
- 전체 합계와 각 구성 요소의 비율을 동시에 파악할 수 있다.
- 항목별 비교와 전체 합계를 한눈에 볼 수 있어 유용하다.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

### 간단한 예제 데이터
- 먼저 간단한 예제로 누적 막대 그래프의 원리를 이해해보자.


```python
# 간단한 예제 데이터
students = ['박태근', '이영희', '박지성']
korean = [85, 90, 78]
english = [88, 85, 92]
math = [90, 88, 85]
```

## 누적 막대 그래프 만들기
### bottom 파라미터
- **bottom** 파라미터를 사용하여 막대를 쌓을 수 있다.
- 첫 번째 막대 위에 두 번째 막대를 쌓으려면 bottom에 첫 번째 데이터를 전달한다.


```python
plt.bar(students, korean)
plt.bar(students, english, bottom=korean) #english는 korean 끝에 쌓아올릴거다.
plt.bar(students, math, bottom= [korean + english for korean, english in zip(korean,english)])

# zip 함수 = 여러개의 순차 자료형 (인덱스가있는자료형= 리스트,튜플)을 병렬로 처리할때 사용
#(korean[0] + english[0]) ,(korean[1] + english[1]), (korean[2] + english[2])...

plt.title('국영수')
plt.ylabel('점수')
plt.show()
```

```python
# zip 함수 = 여러개의 순차 자료형 (인덱스가있는자료형= 리스트,튜플)을 병렬로 처리할때 사용
a= [1,2,3] 
b=[4,5,6]
for i,j in zip(a,b):
    print(i,j)
```

### 세 개 이상 데이터 누적하기
- 세 번째 막대를 쌓으려면 이전 두 데이터의 합을 bottom에 전달한다.


### np.array
- 여러개의 값을 담을 수 있는 다차원 그리드
- 1차원(배열), 2차원(행렬), 3차원 이상(텐서)의 형태를 가질 수 있다.

**특징**
- 안의 모든 요소가 동일한 자료형이어야한다.
- 요소별 연산이 가능하다. (배열전체에 대한 한번의 연산으로)

```python
import numpy as np
scores= np.array([70,85,90])
final_scores = scores + 5
print(final_scores)
```

```python
plt.bar(students, korean, label='국')
plt.bar(students, english, bottom=korean, label='영') #english는 korean 끝에 쌓아올릴거다.
plt.bar(students, math, bottom=np.array(korean)+np.array(english),label='수') #bottom= [korean + english for korean, english in zip(korean,english)
# zip 함수 = 여러개의 순차 자료형 (인덱스가있는자료형= 리스트,튜플)을 병렬로 처리할때 사용
#(korean[0] + english[0]) ,(korean[1] + english[1]), (korean[2] + english[2])...

plt.title('국영수')
plt.ylabel('점수')
plt.legend(fontsize=10)
plt.show()
```

### 범례 추가하기
- 각 데이터에 label을 추가하고 legend()를 호출하면 범례가 표시된다.
